```python
import os
import csv 
import pandas as pd

# Directory for the generated markdown files
output_dir = "./_funding"
os.makedirs(output_dir, exist_ok=True)

# Define grants data (title + excerpt)
grants = pd.read_csv('Xin_Yu_grants.csv').values.tolist()

# Generate markdown files
for i, (title, excerpt, link) in enumerate(grants, 1):
    filename = f"{output_dir}/funding-{i:02d}.md"
    with open(filename, "w") as f:
        f.write(f"---\n")
        f.write(f"title: \"{title}\"\n")
        f.write(f"excerpt: \"{excerpt}\"\n")
        f.write(f"collection: funding\n")
        f.write(f"link: \"{link}\"\n")
        f.write(f"---\n")

output_dir
```
